For this project, you will individually create a program that has a "player" and a "computer" advisary. The computer should randomly choose it's decision based on a list of actions it can take ("Rock", "Paper" or "Scissors"). The player should then have a chance to input their decision. If player and computer choose the same decision then display ("Game Tied"), If the player chooses "Rock" and the computer chooses "Paper" display ("You lose"), if the player chooses "Scissors" and the computer chooses "Rock" display ("You Lose") and if the player chooses "Paper" and the computer chooses "Scissors" then display ("You lose") -- Vice versa for other descisions.

Continue to ask the player for their input until they say "I quit", at which time the game will then end and display ("Thank you for playing").

In this project, you will need to use the random.randint function to enable to computer to make a random decision. Full documentation on the use of this function is attached in a link to this assignment.

```python
#using nested if statements
import random

t = ["rock","paper","scissors"]
compAnswer = random.choice(t)

userWins, compWins, ties = 0,0,0

while True:
    #getting the user's answer
    userAnswer = input("Rock, Paper, Scissors? Enter I quit to finish. Enter check score to see the score ").lower()

    print(userAnswer)
    #All possible outcomes
    paperRock = "Paper covers Rock"
    scissorPaper = "Scissors cut Paper"
    rockScissors = "Rock smashes Scissors"

    #outcome variations
    if userAnswer == compAnswer:
        print("It's a tie. Computer chose:", compAnswer)
        ties+=1
    elif userAnswer == "rock":
        if compAnswer == "paper":
            print("You loose. Computer chose:", compAnswer,'\n', paperRock)
            compWins += 1
        elif compAnswer == "scissors":
            print("You win. Computer chose:", compAnswer,'\n', rockScissors)
            userWins += 1
    elif userAnswer == "paper":
        if compAnswer == "scissors":
            print("You loose. Computer chose:", compAnswer,'\n', scissorPaper)
            compWins += 1
        elif compAnswer == "rock":
            print("You win. Computer chose:", compAnswer,'\n', paperRock)
            userWins += 1
    elif userAnswer == "scissors":
        if compAnswer == "rock":
            print("You loose. Computer chose:", compAnswer,'\n', rockScissors)
            compWins += 1
        elif compAnswer == "paper":
            print("You win. Computer chose:", compAnswer,'\n', paperRock)
            userWins += 1
    elif userAnswer == "check score" or userAnswer == 'check':
        print("Player has", userWins, "points")
        print("Computer has", compWins, "points")
        print('Ties:', ties)
    elif userAnswer == "i quit" or userAnswer == 'quit':
        break
    else:
        print(userAnswer.title(),'isn\'t an option. Please try again')
    compAnswer = random.choice(t)
    
print('Final score: \nYour wins:',userWins,'\nComputer wins:',compWins, '\nTies:', ties)
```

```python
# using a dictionary and if statements
import random

t = ["rock","paper","scissors"]
comp_answer = random.choice(t)

user_wins, comp_wins, tie = 0,0,0

while True:
    user_answer = input("Rock, Paper, Scissors? Enter I quit to finish. Enter check score to see the score ").lower()

    lose_message = f'You loose, computer chose: {comp_answer} \n {comp_answer} beats {user_answer}'
    win_message = f'You win! Computer chose: {comp_answer}\n {user_answer} beats {comp_answer}'
    
    scenario = {
    'rock':{
        'scissors':win_message,
        'paper':lose_message},
    'paper':{
        'rock':win_message,
        'scissors':lose_message},
    'scissors':{
        'paper':win_message,
        'rock':lose_message}}
    
    if user_answer == comp_answer:
        print(f'Tie. Computer chose: {comp_answer}')
        tie += 1
    elif user_answer in scenario:
        print(scenario[user_answer][comp_answer])
        if scenario[user_answer][comp_answer] == win_message:
            user_wins += 1
        elif scenario[user_answer][comp_answer] == lose_message:
            comp_wins += 1
    elif user_answer == 'i quit':
        break
    elif user_answer == 'check score':
        print(f'Player has {user_wins}\nComputer has {comp_wins}\nTies: {tie}')
    else:
        print(user_answer.title(),'isn\'t an option. Please try again')
    comp_answer = random.choice(t)
           
print('Final score: \nYour wins:',user_wins,'\nComputer wins:',comp_wins, '\nTies:', tie)
```
